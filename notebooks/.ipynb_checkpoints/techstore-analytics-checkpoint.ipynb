```python
# Manipulação de dados
import numpy as np
import pandas as pd

# Visualização 
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns

# Garantir reprodutibilidade dos resultados
np.random.seed(42)
```

<hr style="height:3px;background-color:#1F4E79;">

<h1 style="color:#1F4E79;">
2. FUNÇÕES AUXILIARES
</h1>

```python
# Gráfico de barras para as análises
def grafico_barras(dados, titulo='Gráfico de Barras', eixo_x='Categorias', eixo_y='Valores'):
    
    largura = max(8, len(dados) * 0.8)

    plt.figure(figsize=(largura, 5))

    ax = sns.barplot(
        x=dados.index,
        y=dados.values
    )

    ax.bar_label(
        ax.containers[0],
        fmt='%.0f'
    )

    
    plt.title(titulo) #
    plt.xlabel(eixo_x) #
    plt.ylabel(eixo_y) #

    plt.xticks(rotation=45)

    plt.show()

# Formatação dos KPIs do Dashboard
def card_kpi(ax, x, titulo, valor):

    ax.text(
        x,
        0.5,
        f'{titulo}\n{valor}',
        transform=ax.transAxes,
        ha='center',
        va='center',
        fontsize=15,
        weight='bold',
        bbox=dict(
            boxstyle='round,pad=0.8',
            facecolor='white',
            edgecolor='black'
        )
    )

def formato_milhoes(x, pos):
    return f'R$ {x/1_000_000:.1f} Mi'

formatter = FuncFormatter(formato_milhoes)
```

<hr style="height:3px;background-color:#1F4E79;">

<h1 style="color:#1F4E79;">
3. 🏗️ GERAÇÃO DOS DADOS
</h1>

<h2 style="color:#2E75B6;">
3.1 Catálogo de Produtos
</h2>

```python
# ============================================================================
# CATÁLOGO DE PRODUTOS
# ============================================================================
#
# Estrutura contendo:
# - Produto
# - Categoria
# - Preço de Venda
# - Custo Unitário
#
# Será utilizada posteriormente para enriquecer o DataFrame.
# ============================================================================

# Definindo os produtos da Loja e seus respectivos dados: Categoria, Preço e Custo de produção
produtos = {
    # COMPUTADORES
    'Notebook': {'categoria': 'Computadores', 'preco': 7500.00, 'custo': 4795.00},
    'Notebook Gamer': {'categoria': 'Computadores', 'preco': 11999.99, 'custo': 8999.99},

    # MONITORES
    'Monitor Full HD': {'categoria': 'Monitores', 'preco': 2799.99, 'custo': 1499.99},
    'Monitor Gamer': {'categoria': 'Monitores','preco': 4500.00,'custo': 2850.00},

    # PERIFÉRICOS
    'Mouse': {'categoria': 'Periféricos', 'preco': 249.99, 'custo': 79.99},
    'Mouse Gamer': {'categoria': 'Periféricos','preco': 449.99,'custo': 179.99},
    'Teclado': {'categoria': 'Periféricos', 'preco': 499.99, 'custo': 299.99},
    'Teclado Mecânico': {'categoria': 'Periféricos','preco': 899.99,'custo': 450.00},

    # ÁUDIO
    'Headset': {'categoria': 'Áudio', 'preco': 799.99, 'custo': 349.99},
    'Fone Bluetooth': {'categoria': 'Áudio','preco': 649.99,'custo': 279.99},

    # VÍDEO
    'Webcam': {'categoria': 'Vídeo', 'preco': 349.99, 'custo': 179.99},

    # ARMAZENAMENTO
    'SSD': {'categoria': 'Armazenamento', 'preco': 599.99, 'custo': 199.99},
     'HD Externo': {'categoria': 'Armazenamento','preco': 749.99,'custo': 349.99},

    # COMPONENTES
    'Placa de Vídeo': {'categoria': 'Componentes', 'preco': 4500.00, 'custo': 2000.00},
    'Memória RAM': {'categoria': 'Componentes','preco': 699.99,'custo': 299.99
    }
}

# Produtos mais populares possuem maior probabilidade de venda.
#
# Exemplo:
# Mouse e Teclado possuem maior saída.
# Placa de Vídeo possui menor volume,
# porém maior ticket.

# Definindo a probabilidade da ocorrência de cada produto.
pesos_produtos = [
    
    # Computadores
    0.05,  # Notebook
    0.02, # Notebook Gamer

    # Monitores
    0.05, # Monitor Full HD
    0.03, # Monitor Gamer

    # Periféricos
    0.15,  # Mouse
    0.08,  # Mouse Gamer
    0.13,  # Teclado
    0.07,  # Teclado Mecânico

    # Áudio
    0.08,  # Headset
    0.07,  # Fone Bluetooth

    # Vídeo
    0.06, # Webcam

    # Armazenamento
    0.10, #SSD
    0.06, # HD Externo

    # Componentes
    0.03,  # Placa de Vídeo
    0.02   # Memória RAM
]

# Escolhendo aleatoriamente o nome de 1000 produtos e aplicando a probabilidade de cada um ser escolhido
produtos_vendidos = np.random.choice(
    list(produtos.keys()),
         size = 1000,
         p=pesos_produtos
)

# Criando quantidades fictícias de vendas desses produtos
quantidades = np.random.randint(1, 6, size = 1000)
```

<h2 style="color:#2E75B6;">
3.2 Calendário e Sazonalidade
</h2>

```python
# Representação numérica dos meses do ano.
#
# Será utilizada para simular sazonalidade nas vendas.
meses = list(range(1,13))

# Definindo a probabilidade da ocorrência de cada mês
pesos_meses = [
    0.04, # Jan
    0.05, # Fev
    0.06, # Mar
    0.07, # Abr
    0.07, # Mai
    0.08, # Jun
    0.09, # Jul
    0.08, # Ago
    0.08, # Set
    0.10, # Out
    0.15, # Nov
    0.13  # Dez
]

# Dicionário que guarda a chave, valor de cada mês 
nomes_meses = {
    1: 'Janeiro',
    2: 'Fevereiro',
    3: 'Março',
    4: 'Abril',
    5: 'Maio',
    6: 'Junho',
    7: 'Julho',
    8: 'Agosto',
    9: 'Setembro',
    10: 'Outubro',
    11: 'Novembro',
    12: 'Dezembro'
}
```

<h2 style="color:#2E75B6;">
3.3 Vendedores
</h2>

```python
# Criando vendedores fictícios
vendedores = ['Ana', 'Bruno', 'Carlos', 'Luciano', 'Fernanda', 'Juliana', 'Marcos', 'Brendon', 'Eduardo', 'Bianca']

# Criando pesos para aplicar nas vendas de cada vendedor
pesos_vendedores = [
    0.22, # Ana
    0.08, # Bruno
    0.12, # Carlos
    0.10, #Luciano
    0.08, # Fernanda
    0.02, # Juliana
    0.02, # Marcos
    0.10, # Brendon
    0.10, #Eduardo
    0.16 # Bianca
]
```

<h2 style="color:#2E75B6;">
3.4 Regiões
</h2>

```python
# Criando uma lista das regiões do Brasil
regioes = ['Sudeste', 'Sul', 'Nordeste', 'Centro-Oeste', 'Norte']

pesos_regioes = [
    0.45,  # Sudeste
    0.20,  # Sul
    0.18,  # Nordeste
    0.10,  # Centro-Oeste
    0.07   # Norte
]
```

<h2 style="color:#2E75B6;">
3.5 Construção do DataFrame
</h2>

```python
# Criando um DataFrame e inserindo as primeiras colunas de Produto: "produto" e "quantidade"
df = pd.DataFrame({
    'id_venda': range(1, len(produtos_vendidos) + 1),
    'produto': produtos_vendidos,
    'quantidade': quantidades
})

# Coluna índice do mês
df['mes'] = np.random.choice(
    meses, 
    size=len(df),
    p=pesos_meses
)

# Coluna nome no do mês
df['nome_mes'] = df['mes'].map(nomes_meses)

# Coluna vendedor
df['vendedor'] = np.random.choice(
    vendedores, 
    size=len(df),
    p=pesos_vendedores
)

# Coluna regiao
df['regiao'] = np.random.choice(
    regioes,
    size=len(df),
    p=pesos_regioes
)
```

<hr style="height:3px;background-color:#1F4E79;">

<h1 style="color:#1F4E79;">
4. ⚙️ ENGENHARIA DE DADOS
</h1>

<h2 style="color:#2E75B6;">
4.1 Categoria
</h2>

```python

categorias = {produto: dados['categoria'] for produto, dados in produtos.items()}
df['categoria'] = df['produto'].map(categorias) 
```

<h2 style="color:#2E75B6;">
4.2 Preço
</h2>

```python

precos = {produto: dados['preco'] for produto, dados in produtos.items()}
df['preco_unitario'] = df['produto'].map(precos) 
```

<h2 style="color:#2E75B6;">
4.3 Custo Unitário
</h2>

```python

custos = {produto: dados['custo'] for produto, dados in produtos.items()} 
df['custo_unitario'] = df['produto'].map(custos)
```

<h2 style="color:#2E75B6;">
4.4 Faturamento
</h2>

```python
# Receita bruta gerada pela venda (comentário de negócio)
df['faturamento'] = (
    df['preco_unitario'] * df['quantidade']
)
```

<h2 style="color:#2E75B6;">
4.5 Custo total
</h2>

```python
# Custo total associado à venda
df['custo_total'] = (
    df['custo_unitario'] * df['quantidade']
)
```

<h2 style="color:#2E75B6;">
4.6 Lucro
</h2>

```python
# Resultado financeiro da venda
df['lucro'] = (
    df['faturamento'] - df['custo_total']
)
```

<h2 style="color:#2E75B6;">
4.7 Margem
</h2>

```python
df['margem_percentual'] = (
    df['lucro'] / df['faturamento']
) * 100
```

### Quais produtos são responsáveis pela maior parte do faturamento da loja?

```python
#----------------------------------------------------------- CONSULTA

# Análise de pareto
pareto_produto = (
    df.groupby('produto')['faturamento']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

pareto_produto.columns = [
    'produto',
    'faturamento'
]

# participacao especial
pareto_produto['participacao_%'] = (
    pareto_produto['faturamento']
    /
    pareto_produto['faturamento'].sum()
) * 100

# participacao acumulada

pareto_produto['participacao_acumulada_%'] = (
    pareto_produto['participacao_%']
    .cumsum()
)

pareto_produto.head()
```

```python
#----------------------------------------------------------- GRÁFICO

plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=pareto_produto,
    x='produto',
    y='faturamento'
)

plt.xticks(rotation=45)

plt.title('Análise de Pareto - Faturamento por Produto')

# Segundo eixo y
ax2 = plt.twinx()

# Linha da participação acumulada
ax2.plot(
    pareto_produto['produto'],
    pareto_produto['participacao_acumulada_%'],
    marker='o'
)

# valores acumulados
for i, valor in enumerate(pareto_produto['participacao_acumulada_%']):
    # "Desenhe um texto na posição do ponto mostrando seu valor percentual"
    ax2.text(
        i,                # posição horizontal
        valor + 1,        # posição vertical (um pouco acima do ponto)
        f'{valor:.0f}%',  # texto exibido
        ha='center'
)

# Linha dos 80%
ax2.axhline(
    80,
    linestyle='--',
    color='red',
    label='Limite Pareto (80%)'
)

ax2.legend()

ax2.set_ylabel('Participação Acumulada (%)')

plt.show()
```

<div style="
background-color:#D5F5E3;
padding:12px;
border-left:5px solid green;
">

<b>Insights</b><br>

A análise de Pareto revelou uma forte concentração da receita em uma parcela reduzida do portfólio. Dos 15 produtos comercializados pela empresa, apenas 7 foram responsáveis por aproximadamente 80% do faturamento total.

O Notebook destacou-se como principal produto da operação, contribuindo sozinho com cerca de 29% da receita. Em conjunto com Notebook Gamer, Monitor Full HD, Placa de Vídeo, Monitor Gamer, Headset e Teclado Mecânico, forma o grupo de produtos estratégicos para o negócio.

Esse comportamento evidencia que a empresa depende significativamente de um conjunto restrito de itens para sustentar sua geração de receita, característica comum em operações de varejo tecnológico.
</div>

<div style="
background-color:#FCF3CF;
padding:12px;
border-left:5px solid orange;

<b>Recomendação:</b><br>

Os produtos que compõem a faixa dos 80% do faturamento devem receber atenção prioritária em ações comerciais, planejamento de estoque e campanhas promocionais. A indisponibilidade ou queda de desempenho desses itens pode impactar diretamente os resultados financeiros da empresa.
