## Numpy (Computação Numérica)

```python
import numpy as np

#Transforma uma estrutura Python em um array NumPy.
arr = np.array([1,2,3,4,5])
arr * 3
```

## Pandas (Manipulação de Tabelas)

```python
import pandas as pd

alunos = {
    "Nome": ["Eduardo", "Julie", "Guilherme", "Luciano"],
    "Idade": [25, 22, 35, 54] 
}

#Criando uma estrutura tabular
df = pd.DataFrame(alunos)

#COMANDOS
#df.head() -> mostra as primeiras 5 linhas
#df.info() -> mostra informações gerais
#df.describe() -> gera estatísticas
```

---

## Matplotlib (Construção de gráficos)

```python
import matplotlib.pyplot as plt

linhas = [1,2,3,4,5]
colunas = [10, 20, 30, 40, 50]

#Define linhas e colunas
plt.plot(linhas, colunas)

#Mostra o gráfico
plt.show()
```

## Seaborn (Visualizações estatísticas)

```python
import seaborn as sns

#Criando gráfico de barras
sns.barplot(
    data = df,
    x = "Nome",
    y = "Idade"
)

```
